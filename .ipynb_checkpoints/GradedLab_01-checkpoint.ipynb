```python
import numpy as np
import gymnasium as gym
import random
import matplotlib.pyplot as plt

from pathlib import Path
from typing import NamedTuple
from gymnasium.envs.toy_text.frozen_lake import generate_random_map
```

# TASK 1 - Monte Carlo with a random policy

```python
# Define parameters using a NamedTuple
class Params(NamedTuple):
    total_episodes: int
    gamma: float
    map_size: int
    seed: int
    is_slippery: bool
    proba_frozen: float
    savefig_folder: Path

# Initialize parameters
params = Params(5000,
    gamma=0.9,
    map_size=5,
    seed=123,
    is_slippery=True,
    proba_frozen=0.8,
    savefig_folder=Path("./img"),
)

# Create the folder for saving figures if it doesn't exist
params.savefig_folder.mkdir(parents=True, exist_ok=True)

# Initialize the Frozen Lake environment with random map generation
env = gym.make(
    "FrozenLake-v1",
    is_slippery=params.is_slippery,
    render_mode='rgb_array', 
    desc=generate_random_map(
        size=params.map_size,
        p=params.proba_frozen,
        seed=params.seed
    ),
)

# Initialize state-value table and tracking variables
V = np.zeros(env.observation_space.n)
returns_sum = np.zeros(env.observation_space.n)
returns_count = np.zeros(env.observation_space.n)

# Function to generate an episode using a random policy
def generate_episode(env):
    episode = []
    state, _ = env.reset()
    done = False
    while not done:
        action = env.action_space.sample()  # Random action
        next_state, reward, done, _, *_ = env.step(action)
        episode.append((state, action, reward))
        state = next_state
    return episode


# Monte Carlo method with First-Visit
for episode in range(params.total_episodes):
    episode_data = generate_episode(env)
    visited_states = set()
    G = 0
    
    # Calculate returns and update state-value function incrementally
    for t in reversed(range(len(episode_data))):
        state, action, reward = episode_data[t]
        G = params.gamma * G + reward
        if state not in visited_states:
            visited_states.add(state)
            returns_sum[state] += G
            returns_count[state] += 1
            V[state] = returns_sum[state] / returns_count[state]

# Reshape state-value function for visualization
grid_size = int(np.sqrt(env.observation_space.n))
V_grid = V.reshape((grid_size, grid_size))

# Plot the state-value function as a heatmap
plt.figure(figsize=(8, 8))
heatmap = plt.imshow(V_grid, cmap="viridis", origin="upper")
plt.colorbar(heatmap, label="State-Value (V)")
plt.title("State-Value Function (Monte Carlo) - Frozen Lake", fontsize=12)
plt.xlabel("State (Columns)")
plt.ylabel("State (Rows)")

# Display the values on the heatmap
for i in range(grid_size):
    for j in range(grid_size):
        plt.text(j, i, f"{V_grid[i, j]:.2f}", ha="center", va="center", color="black")

plt.xticks(range(grid_size))
plt.yticks(range(grid_size))
plt.show()
```

```python
# Track episode lengths and returns
for episode in range(params.total_episodes):
    episode_data = generate_episode(env)
    episode_lengths.append(len(episode_data))  # Track trajectory length
    G = 0
    visited_states = set()
    for t in reversed(range(len(episode_data))):
        state, action, reward = episode_data[t]
        G = params.gamma * G + reward
        if state not in visited_states:
            visited_states.add(state)
            returns_sum[state] += G
            returns_count[state] += 1
            V[state] = returns_sum[state] / returns_count[state]
    episode_returns.append(G)  # Track return for learning curve
```

# TASK 2 - Incremental Monte Carlo

```python
# Define parameters using a NamedTuple
class Params(NamedTuple):
    total_episodes: int
    gamma: float
    map_size: int
    seed: int
    is_slippery: bool
    proba_frozen: float
    savefig_folder: Path

# Initialize parameters
params = Params(
    total_episodes=5000,
    gamma=0.9,
    map_size=5,
    seed=123,
    is_slippery=True,
    proba_frozen=0.8,
    savefig_folder=Path("./img"),
)

# Create the folder for saving figures if it doesn't exist
params.savefig_folder.mkdir(parents=True, exist_ok=True)

# Initialize the Frozen Lake environment with random map generation
env = gym.make(
    "FrozenLake-v1",
    is_slippery=params.is_slippery,
    render_mode='rgb_array', 
    desc=generate_random_map(
        size=params.map_size,
        p=params.proba_frozen,
        seed=params.seed
    ),
)

# Initialize state-value table and tracking variables
V = np.zeros(env.observation_space.n)
returns_sum = np.zeros(env.observation_space.n)
returns_count = np.zeros(env.observation_space.n)

# Function to generate an episode using a random policy
def generate_episode():
    episode = []
    state, _ = env.reset()
    done = False
    while not done:
        action = env.action_space.sample()  # Random action
        next_state, reward, done, _, *_ = env.step(action)
        episode.append((state, action, reward))
        state = next_state
    return episode

# Incremental Monte Carlo
for episode in range(params.total_episodes):
    episode_data = generate_episode()
    G = 0
    
    # Loop through the episode in reverse for incremental updates
    for t in reversed(range(len(episode_data))):
        state, action, reward = episode_data[t]
        G = params.gamma * G + reward  # Compute return
        
        # Incremental update of state-value function
        V[state] += (1 / (returns_count[state] + 1)) * (G - V[state])
        returns_count[state] += 1

# Reshape state-value function for visualization
grid_size = int(np.sqrt(env.observation_space.n))
V_grid = V.reshape((grid_size, grid_size))

# Plot the state-value function as a heatmap
plt.figure(figsize=(8, 8))
heatmap = plt.imshow(V_grid, cmap="viridis", origin="upper")
plt.colorbar(heatmap, label="State-Value (V)")
plt.title("State-Value Function (Incremental Monte Carlo) - Frozen Lake", fontsize=12)
plt.xlabel("State (Columns)")
plt.ylabel("State (Rows)")

# Display the values on the heatmap
for i in range(grid_size):
    for j in range(grid_size):
        plt.text(j, i, f"{V_grid[i, j]:.2f}", ha="center", va="center", color="black")

plt.xticks(range(grid_size))
plt.yticks(range(grid_size))
plt.show()
```

```python
# Track episode lengths and returns
for episode in range(params.total_episodes):
    episode_data = generate_episode(env)
    episode_lengths.append(len(episode_data))  # Track trajectory length
    G = 0
    for t in reversed(range(len(episode_data))):
        state, action, reward = episode_data[t]
        G = params.gamma * G + reward
        V[state] += (1 / (returns_count[state] + 1)) * (G - V[state])
        returns_count[state] += 1
    episode_returns.append(G)  # Track return for learning curve
```

# TASK 3 - Q-Learning Integration

```python
import numpy as np
import gymnasium as gym
import matplotlib.pyplot as plt

# Parameters
n_episodes = 5000       # Number of episodes
gamma = 0.9             # Discount factor
alpha = 0.1             # Learning rate
epsilon = 1.0           # Initial exploration rate
min_epsilon = 0.1       # Minimum exploration rate
epsilon_decay = 0.995   # Decay rate for epsilon

# Initialize the Frozen Lake environment
env = gym.make("FrozenLake-v1", is_slippery=True)

# Initialize Q-table to zeros
Q = np.zeros((env.observation_space.n, env.action_space.n))

# Function to choose action using -Greedy policy based on Q-values
def choose_action(state):
    if np.random.uniform(0, 1) < epsilon:
        return env.action_space.sample()  # Explore: random action
    else:
        return np.argmax(Q[state])  # Exploit: action with max Q-value for the state

# Q-Learning algorithm
for episode in range(n_episodes):
    state, _ = env.reset()
    done = False

    while not done:
        # Choose action with -Greedy policy
        action = choose_action(state)
        next_state, reward, done, _, *_ = env.step(action)

        # Q-Learning update
        best_next_action = np.argmax(Q[next_state])  # Best action at next state
        Q[state, action] += alpha * (reward + gamma * Q[next_state, best_next_action] - Q[state, action])

        state = next_state  # Move to the next state

    # Decay  to reduce exploration over time
    epsilon = max(min_epsilon, epsilon * epsilon_decay)

# Display learned Q-table
print("Learned Q-Table:")
print(Q)

# Derive and display the policy from Q-table
policy = np.argmax(Q, axis=1)
print("Derived Policy (best action for each state):")
print(policy.reshape((int(np.sqrt(env.observation_space.n)), -1)))

# Visualize the State-Value Function (derived from Q-table)
V = np.max(Q, axis=1)  # Max Q-value for each state to represent the state-value
grid_size = int(np.sqrt(env.observation_space.n))
V_grid = V.reshape((grid_size, grid_size))

# Plot the state-value function as a heatmap
plt.figure(figsize=(8, 8))
heatmap = plt.imshow(V_grid, cmap="viridis", origin="upper")
plt.colorbar(heatmap, label="State-Value (V)")
plt.title("State-Value Function (Q-Learning) - Frozen Lake", fontsize=12)
plt.xlabel("State (Columns)")
plt.ylabel("State (Rows)")

# Display the values on the heatmap
for i in range(grid_size):
    for j in range(grid_size):
        plt.text(j, i, f"{V_grid[i, j]:.2f}", ha="center", va="center", color="black")

plt.xticks(range(grid_size))
plt.yticks(range(grid_size))
plt.show()
```

```python
# Q-Learning algorithm
for episode in range(params.total_episodes):
    state, _ = env.reset()
    done = False
    episode_length = 0  # Track the length of each episode

    while not done:
        action = choose_action(state)
        next_state, reward, done, _, *_ = env.step(action)
        
        # Q-Learning update
        best_next_action = np.argmax(Q[next_state])  # Best action at next state
        Q[state, action] += alpha * (reward + gamma * Q[next_state, best_next_action] - Q[state, action])

        state = next_state  # Move to the next state
        episode_length += 1  # Increment the episode length

    episode_lengths.append(episode_length)  # Track trajectory length
    episode_returns.append(reward)  # Track the last reward (return for this episode)

    # Decay  to reduce exploration over time
    epsilon = max(min_epsilon, epsilon * epsilon_decay)
```

```python
# Plot Trajectory Lengths
plt.figure(figsize=(10, 4))
plt.plot(episode_lengths)
plt.title("Episode Lengths over Time")
plt.xlabel("Episode")
plt.ylabel("Length of Trajectory")
plt.show()

# Plot Learning Curve (Returns over Episodes)
plt.figure(figsize=(10, 4))
plt.plot(episode_returns)
plt.title("Learning Curve: Return per Episode")
plt.xlabel("Episode")
plt.ylabel("Return")
plt.show()
```
